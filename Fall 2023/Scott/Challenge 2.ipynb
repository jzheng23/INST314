```python
import numpy as np
from scipy.stats import norm, uniform
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# Part 1

1. generate 10000 values from normal distribution

```python
normal_sample = norm.rvs(size = 10000) 
```

2. Plot a histogram of the random values

```python
sns.histplot(normal_sample, bins = 30, edgecolor=None)
```

3.  Generate a large number of values (the same that you settled on for from a uniform distribution between -3 and 3.

```python
uniform_sample = uniform.rvs(size = 10000) *6 -3
```

4. histogram of the uniform sample

```python
sns.histplot(uniform_sample, bins = 30, edgecolor=None)
```

5. generate the "grid" of data values by
    creating the sequence of numbers from -4 to 4, incrementing by 0.1.

```python
x_grid = np.linspace(-4, 4, 81)
norm_densities = norm.pdf(x_grid)
```

6. Create a line plot

```python
sns.lineplot(x=x_grid, y=norm_densities)
```

7. compute probability densities from a
    uniform distribution that goes from -3 to 3, but use the same "grid"

```python
unif_densities = uniform.pdf(x_grid, loc=-3, scale=6)
```

8.  Plot the uniform densities

```python
sns.lineplot(x=x_grid, y=unif_densities)
```

9. Generate a small number of random numbers from a normal
    distribution

```python
small_norm = norm.rvs(size = 10)
```

10. Create a combined plot

```python
sns.histplot(small_norm, bins = 30, edgecolor=None, kde = True)
```

11. generate a small
    number of random numbers from the same uniform distribution

```python
small_uniform = uniform.rvs(size = 10) *6 -3
```

12. combined plot

```python
sns.histplot(small_uniform, bins = 30, edgecolor=None, kde = True)
```

    distribution.

```python
all_rolls = [sum(np.random.choice(20, 30) + 1) 
             for i in range(1000)]
```

31. density values of normal curve

```python
x_grid = np.linspace(np.min(glucose), np.max(glucose), 100)
norm_densities = norm.pdf(x_grid, 
                          loc = np.mean(glucose), 
                          scale = np.std(glucose))
```

32. normal curve alone

```python
sns.lineplot(x= x_grid,y = norm_densities)
```

33. hist with normal curve

```python
sns.histplot(x = glucose,  stat="density", bins = 30, edgecolor = None)
sns.lineplot(x= x_grid,y = norm_densities)
```
